```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
total = []
r_val = np.linspace(1,4,100000)
for r in r_val:
    x = np.random.random()
    arr = [x]
    for i in range(1,101):
        arr.append(r*arr[i-1]*(1-arr[i-1]))
        if(i==100):
            total.append(arr[i])
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(np.arange(min(r_val), max(r_val)+1, 1.0))
plt.plot(r_val,total)
```

```python
plt.figure(figsize=(50, 40))
plt.plot(r_val,total,ls='', marker=',')
```

```python
l_total = []
for r in r_val:
    arr = [0.2]
    sum = 0
    for i in range(1,100):
        arr.append(r*arr[i-1]*(1-arr[i-1]))
        sum+=math.log(abs(r-2*arr[i-1]*r))
    sum/=100
    l_total.append(sum)
```

```python
plt.figure(figsize=(50, 40))
plt.plot(r_val,l_total)
```

```python
total_1 = []
for r in r_val:
    x = np.random.random()
    arr = [x]
    for i in range(1,101):
        arr.append(r*math.sin(math.pi*arr[i-1]))
        if(i==100):
            total_1.append(arr[i])
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(np.arange(min(r_val), max(r_val)+1, 1.0))
plt.plot(r_val,total_1)
```

```python
plt.figure(figsize=(50, 40))
plt.plot(r_val,total_1,ls='', marker=',')
```
