```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
y_axis = []
x_axis = []
lyapunov_plot = []
r_val = np.linspace(1,4,10000)
for r in r_val:
    to_plot = []
    x = 0.2
    lyapunov = 0
    for i in range(500):
        xi = abs(r - 2*x*r)
        lyapunov+=math.log(xi)
        x = r*x*(1-x)
        if(i>=450):
            to_plot.append(x)
    lyapunov_plot.append(lyapunov/500)
    y_axis.extend(to_plot)
    for i in range(50):
        x_axis.append(r)
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(np.arange(min(r_val), max(r_val)+1, 1.0))
plt.plot(x_axis,y_axis)
```

```python
plt.figure(figsize=(30, 20))
plt.plot(x_axis,y_axis,ls='', marker=',')
```

```python
plt.figure(figsize=(60, 400))
plt.plot(x_axis,y_axis,ls='', marker=',')
plt.plot(r_val,lyapunov_plot,'r')
```

```python
total_1 = []
for r in r_val:
    x = np.random.random()
    arr = [x]
    for i in range(1,101):
        arr.append(r*math.sin(math.pi*arr[i-1]))
        if(i==100):
            total_1.append(arr[i])
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(np.arange(min(r_val), max(r_val)+1, 1.0))
plt.plot(r_val,total_1)
```

```python
plt.figure(figsize=(50, 40))
plt.plot(r_val,total_1,ls='', marker=',')
```

```python
def logistic(r, x):
    return r * x * (1 - x)
n = 10000
r = np.linspace(2.5, 4.0, n)
```

```python
iterations = 1000
last = 100
x = 1e-5 * np.ones(n)
```

```python
for i in range(iterations):
    x = logistic(r, x)
    # We compute the partial sum of the
    # Lyapunov exponent.
    # We display the bifurcation diagram.
    if i >= (iterations - last):
        ax1.plot(r, x, ',k', alpha=.25)
        
plt.show()
```
