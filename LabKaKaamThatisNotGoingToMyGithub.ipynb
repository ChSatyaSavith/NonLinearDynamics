```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = [0]
r = 4
x = [0.1]
for i in range(1,51):
    n.append(i)
    x.append(x[i-1]*r*(1-x[i-1]))
```

```python
n1 = [0]
r = 4
x1 = [0.101]
for i in range(1,51):
    n1.append(i)
    x1.append(x1[i-1]*r*(1-x1[i-1]))
```

```python
plt.figure(figsize=(40,20))
plt.plot(n,x,'o-r')
plt.plot(n1,x1,'o-b')
```
