```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
y_axis = []
x_axis = []
lyapunov_plot = []
r_val = np.linspace(0,1,10000)
for r in r_val:
    to_plot = []
    x = 0.2
    lyapunov = 0
    for i in range(500):
#         xi = abs(math.sin(math.pi*r))
#         lyapunov+=math.log(xi)
#         arr.append(r*math.sin(math.pi*arr[i-1]))
        x = r*math.sin(math.pi*x)
        if(i>=450):
            to_plot.append(x)
    lyapunov_plot.append(lyapunov/500)
    y_axis.extend(to_plot)
    for i in range(50):
        x_axis.append(r)
```

```python
plt.figure(figsize=(20, 10))
plt.xticks(np.arange(min(r_val), max(r_val)+1, 1.0))
plt.plot(x_axis,y_axis)
```
